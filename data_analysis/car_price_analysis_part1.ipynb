## 0) Dowload data 

```python
import pandas as pd
import numpy as np
# save the url path, before any action 
# check what it is
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
# create a dataframe object
df = pd.read_csv(url,header=None)

df.head(10)
```

Numbered headers are not exactly readble. We will change them as in information about the table provided at <br>
https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names

```python
# taken headers names
features_names = "symboling:          normalized-losses: make:                   fuel-type:         aspiration:        num-of-doors:      body-style:        drive-wheels:      engine-location:   wheel-base:        length:            width:             height:            curb-weight:       engine-type:       num-of-cylinders:  engine-size:       fuel-system:       bore:              stroke:            compression-ratio: horsepower:        peak-rpm:          city-mpg:          highway-mpg:       price:          "
# we filter unused spaces and split words to a list
features_list = features_names.split()
features_list = [el[:-1] for el in features_list]
# print(features_list)
#print(len(features_list))
# there is 26 features, so evertyhink went fine
headers = features_list 
df.columns = headers
df.head(5)
```

When everything is set up we save our table.
