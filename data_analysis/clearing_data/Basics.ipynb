We can agry that nameing features (clomuns) with nimbers $1, 2, 3...$ is not a readable idea. So let us change names of columns.
<br>
I take the features_names form <br> https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names <br> and transform it a bit so it feets headres names.

```python
features_names = "symboling:          normalized-losses: make:                   fuel-type:         aspiration:        num-of-doors:      body-style:        drive-wheels:      engine-location:   wheel-base:        length:            width:             height:            curb-weight:       engine-type:       num-of-cylinders:  engine-size:       fuel-system:       bore:              stroke:            compression-ratio: horsepower:        peak-rpm:          city-mpg:          highway-mpg:       price:          "
# we filter unused spaces and split words to a list
features_list = features_names.split()
features_list = [el[:-1] for el in features_list]
# print(features_list)
#print(len(features_list))
# there is 26 features, so evertyhink went fine
headers = features_list 
df.columns = headers
df.head(5)
```

If our data is a last as we wanted it to be, we can save our progres:

```python
# paste a desired path
#path = "your path"
#df.to_csv('path')
"""
i do not want to have inexing numbers 
so with index=False we make sure
that those won't appear
""" 
df.to_csv("autos.csv", index=False)
```
