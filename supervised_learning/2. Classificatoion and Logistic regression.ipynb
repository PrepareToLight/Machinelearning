# Classificatoion and Logistic regression
In classification problem we are trying to predict the $y$-labels value based o given data, except (the regression problem) this time they are $\underline{discrete}$. E.g. $y \in \{-1,1\}$, which may represent the customer that liked the product or didn't.

## a) Logistic regression
Let's asume that our classification is $y \in \{0,1\}$.
This time we would like to choose such a hypothesis function that does not take larger nor lower values than 0 and 1. So our hypothesis is <br> 
<center>$h_\theta(x) = \frac{1}{1 + e^{-\theta^T x}}$</center>
If we take $z = \theta^T x$, than 
<center>$h_\theta(x) = g(z) = \frac{1}{1 + e^{-z}}$</center>
still $\theta^T x = \theta_0 + \sum_{i=1}^{d} \theta_i x_i$ <br>
As an exercise try to find $\frac{d}{dz}g$ and express it using $g(z)$.

We have our model that seems to have desired properties, but how do we find the $\theta$ that satisfies most our data? <br>
Let us make some probabilistic assumptions <br>
<center>$\mathbb{P}(y=1|x,\theta) = h_\theta (x)$</center>
<center>$\mathbb{P}(y=0|x,\theta) = 1-h_\theta (x)$</center>
<center></center>
We can for simplicity write it as
<center>$p(y|x,\theta) = (h_\theta(x))^y \cdot ((1-h_\theta)(x))^{1-y}$</center>
Does it seem famirial? <br>
Let's try to find the optimum $\theta$ via maximum likelihood function. We asume that $n$ of samples form our data set $X$ are iid, so we have
<center>$L(\theta) = p(\vec{y} |X,\theta) = \prod_{i=1}^n (h_\theta(x^{(i)}))^{y^{(i)}} \cdot ((1-h_\theta)(x^{(i)}))^{1-y^{(i)}}$</center>
The best way to find maximum of $L$ is to rewrite it as $l(\theta) = \log L(\theta) = \sum_{i=1}^n \log (h_\theta(x^{(i)}))^{y^{(i)}} + \log ((1-h_\theta)(x^{(i)}))^{1-y^{(i)}}$. This way it's a lot easier. <br> To maximize it we will use the gradient ascent 
<center>$\theta \leftarrow \theta + \alpha \nabla l(\theta)$ </center>
The difrence form gradient descent is plus sing :) If we look at $j$th partial derivative for one training  example and solve it. We surprisingly get
<center>$\frac{\partial}{\partial \theta_j} l({\theta}) = (y-h_\theta (x))x_j$</center>
This give us the Stochastic Gradient Ascent (SGA) rule, where we update
<center>$\theta \leftarrow \theta + \alpha (y^{(i)}-h_\theta (x^{(i)}))x^{(i)} $</center>

```python
#toy_features, toy_labels = toy_data = utils.load_toy_data('toy_data.tsv')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
labels = ["labels","feature0", "feature1"]
df = pd.read_csv("toy_data.tsv", sep='\t')
df.columns = labels
df
```

We will implement the SGA on above training set and see how presice it can be. As we already seen the SGA is very similar to GDA

```python
# we implement SGA
# inputs must by np.arrays
def sga(data,labels,grad_fn,alpha,iter_grad):
    N = data.shape[0]
    x0 = np.ones(N)
    # 1st we must asure that x's in data
    # are this x = (1,x1,x2,...,xd)
    data = np.column_stack((x0,data))
    N, D = data.shape
    # dimension of theta must be equal to dim. of x
    # initialize theta
    theta = data[0]
    theta_list = [data[0]]
    for t in range(iter_grad):
        for i in range(N):
            # we make an udate each step
            theta += alpha*grad_fn(data[i],labels[i],theta)
            theta_list.append(theta.copy())
    return np.array(theta_list)
# we difine our learning hypothesis and its gradient
# h_lreg = 1/(1+exp(-z))
h_lreg = lambda x,theta:1/(1+np.exp(-np.dot(x,theta)))
grad_h_lreg = lambda x,y,theta: (y-h_lreg(x,theta))*x
#print(X.shape)
theta_learnd = sga(X,y,grad_h_lreg,0.01,1000)
#print(theta_learnd)
```

```python
fig2, ax2 = plt.subplots(figsize=(10,10))
ax2.plot(theta_learnd[:,0],label=r"$\theta_0$")
ax2.plot(theta_learnd[:,1],label=r"$\theta_1$")
ax2.plot(theta_learnd[:,2],label=r"$\theta_2$")

ax2.set_title(r"Learning process of $\theta$")
ax2.legend()
```

Our learning process is converging, its a good sign, it means our theta is trying to be the most optimate in respect to given training set. Let see how good it will work on it. This time, for simplicity above data set is also a test set.

```python
Data_set = np.column_stack((x0,x1,x2))
trained_y = np.array([h_lreg(x,theta_learnd[-1]) for x in Data_set])
trained_labels = trained_y
trained_y = np.round(trained_y)
trained_y[trained_y == 0] = -1
trained_y
```

Let us look what $\theta$ in our example represents. We have
$\theta^T x = \theta_0 + \theta_1 x_1 + \theta_2 x_2$, if $\theta^T x = 0$ than
<center>$\theta_0 + \theta_1 x_1 + \theta_2 x_2 = 0$</center>
<center>$x_2 = -\frac{\theta_1}{\theta_2}x_1 -\frac{\theta_0}{\theta_2}$</center>
It is a linear function such as $y = ax + b$. Let's know compare those ours two plots.

```python
# defining our linear function in respect of theta
def lin_func(theta, theta0, a, b):
        if theta[1] == 0:
            return -theta0/theta[0], 0
        x = np.linspace(a,b,100)
        y = -(theta[0]/theta[1])*x - theta0/theta[1]
        return x,y
# defining multiple plot fuction

def plot_3plots(x1,x2,theta,labels,t_labels,raw_trained_labels,a=-2,b=4,size=5):
    arguments,values=lin_func(theta[-1][1:],theta[-1][0],a,b)
    fig, (ax0, ax1, ax2) = plt.subplots(nrows = 1, ncols = 3,figsize=(3*size,size))
    ax0.scatter(x1[labels==-1],x2[labels==-1],c='b',label="y=-1")
    ax0.scatter(x1[labels==1],x2[labels==1],c='r',label="y=1")
    ax0.plot(arguments,values)
    ax0.legend()
    # naming axes
    ax0.set_xlabel('x1 feature')
    ax0.set_ylabel('x2 feature')
    ax0.set_title('Scatter plot with respect to labels')
    ax1.scatter(x1[t_labels==-1],x2[t_labels==-1],c='b',label="y=-1")
    ax1.scatter(x1[t_labels==1],x2[t_labels==1],c='r',label="y=1")
    ax1.plot(arguments,values)
    ax1.legend()
    # naming axes
    ax1.set_xlabel('x1 feature')
    ax1.set_ylabel('x2 feature')
    ax1.set_title('Scatter plot with respect to trained labels')
    ax2.scatter(x1,x2,c=raw_trained_labels)
    ax2.plot(arguments,values)
```

<center>$$g(z) =   \begin{cases}
    1 & \text{if } z \geq 0 \\
    0 & \text{if } z < 0
  \end{cases}$$</center>
  where $g(z) = g(\theta^T x) = h_\theta (x)$. Then we deal with $\underline{Perceptron}$ algorithm.


```python
# we implement the perceptron alg
# using previously coded fucntions
h_g = lambda x,theta:1 if np.dot(x,theta) >=0 else 0
grad_h_g = lambda x,y,theta: (y-h_g(x,theta))*x
perceptron_theta = sga(X,y,grad_h_g,0.01,1000)

trained_p_y = np.array([h_g(x,perceptron_theta[-1]) for x in Data_set])
trained_p_y_1 = trained_p_y.copy()
trained_p_y_1[trained_p_y_1 == 0] = -1
arguments_p, values_p = lin_func(perceptron_theta[-1][1:],perceptron_theta[-1][0],-2,4)

plot_3plots(x1,x2,perceptron_theta,y,trained_p_y_1,trained_p_y)

```

## c) Newton's method
Let us imagine we have continous function $f(\theta)$, $\theta \in \mathbb{R}$. If it has a zero value at $\theta_0$ than for some $\theta_1$ near $\theta_0$ we can find a derivative near this point, hence
$f'(\theta_1) = \frac{d}{d\theta}f(\theta_1) \approx \frac{f(\theta_0) - f(\theta_1)}{\theta_0 - \theta_1 } = \frac{f(\theta_1)}{\theta_1 - \theta_0 }$, so
<center>$f'(\theta_1) (\theta_1 - \theta_0) \approx f(\theta_1)$</center>
<center>$  \theta_0 \approx \theta_1 - \frac{f(\theta_1)}{f'(\theta_1)}$</center>
With this simple fact we can implement very usefull algoritm, when we update $\theta$ towards zero point of a function. The Newton's method can be rewiriten as
<center>$\theta \leftarrow \theta -\frac{f(\theta)}{f'(\theta)}$</center>
This method can be generalized to a higher dimensioned functions, such as our hypothesis $h_\theta$. For example if we took out previos logarrithmic maximum likliehood $l(\theta)$, than we could find it's maximum by iterated updates (simplified)
<center>$\theta \leftarrow \theta -\frac{l'(\theta)}{l''(\theta)}$</center>
Generalized version is called Newton-Raphson method. This is a very good method when our data set is not very big and don't have too much points of freedom. Because of Hesian-matrix that we need to caclulate for $l'$, $l''$. Yet this algorithm converges faster even than BGD and recuries less iterations but becasue of Hesian-matrix it may be operations consuming.
